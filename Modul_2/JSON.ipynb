```python
import json
```

```python
url = "https://www.bibsonomy.org/json/search/Bibliothek?items=1000&duplicates=merged"
```

```python
import requests
```

```python
result = requests.get(url)
```

```python
data = result.json()
```

```python
type(data)
```

```python
from pprint import pprint
pprint(data)
```

```python
edited_items = data["items"]
```

```python
pprint(edited_items[:3])
```

```python
edited_items = []

# Auswahl der Informationen, die ausgewählt werden sollen:

desired_entries = ['isbn', 'pub-type', 'label', 'author', 'publisher', 'series', 'editor']

# Nun können wir die Einträge der eingelesenen Datei iterieren und filtern.ür jedes Element im eingelesenen JSON file... (über data.keys() lässt sich leicht herausfinden, wo die relevanten Daten stecken):
for item in edited_items:
        item['name'] = item.pop('label')
        item['id'] = item('isbn')
        item['cat'] = item.pop('pub-type')
        
for item in data["items"]:

    # erstelle ein neues leeres dictionary für die Daten des Elements

    item_export = {}
    
    for entry in desired_entries:

            # ... überprüfe, ob es einen Eintrag gibt (um Fehler zu vermeiden)
    
            if entry in item.keys():

                # füge den Eintrag den Daten zum Element hinzu

                item_export[entry] = item[entry]
            else:
                item_export[entry] = ''
    

        # am Ende: füge den vollständigen (gefilterten) Eintrag der zu exportierenden Liste hinzu

    edited_items.append(item_export)
```

```python
pprint(edited_items)
```

```python
with open("bibsonomy.json", "w") as outfile:
    json.dump(edited_items, outfile, indent=4)
```

```python
with open("bibsonomy.json", "r") as file:
    data = json.load(file)
```
